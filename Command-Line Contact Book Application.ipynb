```python
import os

FILE_NAME = "contacts.txt"

def add_contact():
    name = input("Enter Name: ").strip()
    phone = input("Enter Phone Number: ").strip()
    email = input("Enter Email: ").strip()

    with open(FILE_NAME, "a") as file:
        file.write(f"{name} | {phone} | {email}\n")

    print("\n‚úÖ Contact added successfully!\n")


def view_contacts():
    if not os.path.exists(FILE_NAME):
        print("\n‚ùå No contacts found.\n")
        return

    print("\nüìí Contact List:\n")
    with open(FILE_NAME, "r") as file:
        contacts = file.readlines()

        if not contacts:
            print("No contacts available.\n")
            return

        for contact in contacts:
            print(contact.strip())
    print()


def search_contact():
    keyword = input("Enter name or phone to search: ").lower()

    if not os.path.exists(FILE_NAME):
        print("\n‚ùå No contacts found.\n")
        return

    found = False
    with open(FILE_NAME, "r") as file:
        for line in file:
            if keyword in line.lower():
                print("\nüîç Contact Found:")
                print(line.strip())
                found = True

    if not found:
        print("\n‚ùå No matching contact found.\n")
    else:
        print()


def main_menu():
    while True:
        print("==== Contact Book ====")
        print("1. Add Contact")
        print("2. View Contacts")
        print("3. Search Contact")
        print("4. Exit")

        choice = input("Choose an option (1-4): ").strip()

        if choice == "1":
            add_contact()
        elif choice == "2":
            view_contacts()
        elif choice == "3":
            search_contact()
        elif choice == "4":
            print("\nüëã Exiting Contact Book. Goodbye!")
            break
        else:
            print("\n‚ùå Invalid choice. Try again.\n")


if __name__ == "__main__":
    main_menu()
```
